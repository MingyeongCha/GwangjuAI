####__사용되는 기술__   
#####1.통계기반 기계 번역(SMT, Statistical Machine Translation)   
통계기반 기계번역은 다른 언어로 번역할 때 번역하고자 하는 문장을 단어 혹은 구절 단위로 분할하고 실제 번역되는 언어에서 더 빈번하게 사용되는 단위를 확률로 나타내고, 더 큰 확률의 단위에 가중치를 두어 번역. 이후 번역된 단위를 합칠 때에도 가능한 여러 조합을 번역될 언어로 쓰인 문장과 비교함으로써 가장 유사성이 높은 문장을 선택해 번역. 즉, 한국어 입력 문장이 들어오면, 각 단어가 어떤 영어 단어로 변역될 확률이 얼마인지 데이터 테이블에서 읽어와 가장 확률 높은 단어를 찾아 번역하는 방식.   
다만, 정확성이 떨어지고, 두 언어의 어순이 다른 경우 각 단위를 순서에 맞춰 배열하기 어려움이 있음.   

#####2.인공 신경망 기계 번역(NMT, Neural Machine Translation)    
2014년 조경현 교수가 발표한 논문에서 제안한 RNN 인코더-디코더(RNN Encoder-Decoder)는 스스로 새로운 기계 번역 기술로 거듭나며 인공 신경망 기반 기계 번역(Neural Based Machine Translation)의 시작을 알림.
RNN은 일렬로 나타나는 데이터에서 패턴을 인식하는데 탁월한 성능을 보임
기계 번역을 수행하기 위해서는 문장을 구성하는 단위를 벡터로 표현하는 과정을 거치며 이는 임베딩(Embedding), 혹은 인코딩(Encoding)으로 불림. 문장 단위를 적절하게 임베딩하면 문장 단위 사이의 관계를 모델링할 수 있고, 이는 더 좋은 번역 성능으로 이어짐. 다만, 훈련하는데 시간과 자원이 많이 소요 되며, 자주사용되지 않는 단어를 번역하는데는 취약함 그러나 최근에는 기계 번역기 스스로 학습해 자신의 정학도를 향상하는데 역 번역과 디노이징기술을 이용해 정확도를 개선하고 있음.
#####3.SMT가NMT로 바뀌면 달라지는 점   
기존 SMT는 구절 단위로 변역해 문장을 만들었다면, NMT문장에 담긴 모든 정보, 네트워크를 통째로 학습시켜 번역하는 방법으로 입력 문장이 들어오면, 이 문장을 가상공간의 하나의 점을 의미하는 벡터(좌표값)와 같은 숫자로 변환하고, 학습 데이터를 통해 배운 가중치를 더해 번역을 수행함. 그러므로 한번도 학습한 적이 없는 문장을 입력하더라도 네트워크를 통과시켜 자동으로 번역함

####__사용되는 기술__   

#####1. 자율주행 자동차의 발전 과정   
미국 자동차기술자협회(SAE)는 자율주행 발전을 6단계로 제시   
  - 0단계 : 전통적 주행 :운전자가 모든 것을 통제
  - 1단계 : 부분 보조 주행 : 속도 및 차간거리 유지, 차선 유지등 시스템이 일정 부분 개입
  - 2단계 : 보조 주행 :특정 상황에서 일정 시간동안 보조 주행, 필요시 운전자가 즉시 개입
  - 3단계 : 부분 자율주행: 고속도로와 같은 조건에서 자율 주행, 필요시 운전자가 즉시 개입
  - 4단계 : 고도 자율주행: 제한 상황을 제외한 대부분의 도로에서 자율주행
  - 5단계 : 완전 자율주행: 탑승자는 목적지만 입력, 운전대와 폐달 제거   

#####2. 자율주행 자동차의 구조   
자율주행 자동차의 구조는 크게 센서, 프로세서, 알고리즘, 액추에이터로 구성되어 있음.
센서가 데이터를 수집하면 프로세서가 그 데이터를 처리하고, 알고리즘이 처리 결과를 해석, 주행경로 등 운행에 대해 결정을 내리고 액추에이터가 운행에 필요한 시스템을 제어함.
즉, 자세히 말하면 자율주행 알고리즘은 라이다, 레이더, 센서, 카메라 등의 장치로부터 입력된 주변 상황 정보 및 운전자의 상태등을 분석 하여 가속, 브레이크 작동, 차선 유지 등의 액추에어터가 수행하게끔 하는 것.   
#####3. 자율주행 자동차용 인공지능 시스템   
라이다센서, 레이더, 초음파 센서, 적외선카메라, 모노/스테레오카메라 등을 내외부에 설치하고 이러한 센서 등을 통해 외부 정보를 수집한 후 자동차 스스로 주변 환경을 인식해 위험한 상황을 판단하고 주행경로를 계획하는 등 운전자의 주행 조작을 최소화하며, 스스로 안전한 주행이 가능하도록 함.
자율주행 자동차용 인공지능 시스템은 라이다센서, 카메라 등으로 측정한 데이터와 이미지를 활용한 이미지 처리를 통해 주행 환경상의 다양한 대상물체(장애물)의 거리 및 부피를 측정해 대상 물체와의 정확한 거리를 측정함. 또한 시각, 청각, 촉각 등 다양한 인터페이스를 통해 운전자의 상태 및 감성, 성향 및 차량의 주변상황을 종합적으로 판단하고, 운전자의 음성을 인식하기 위한 음성 인식 분야 등에서 자율주행에 도움을 주기 위한 필수 도구로 활용되고 있음.   
#####4. 현재의 자율주행 자동차   
- 외부 센서에서 수집한 정보를 분석해 핸들조작, 페달작동을 통해 가속, 브레이크 작동, 차선유지, 앞차와의 안전거리 유지 및 주차등에 활용(1~2Level)하고 있음.
- Level3~4 수준의 자율주행자동차는 테스트 주행 중에 있음.

#####1. 안면인식 방법
얼굴의 특징점을 추출하여 저장된 데이터베이스 내 자료와 비교하여 신원을 확인함.
이 때 얼굴의 특징점을 추출하는 방법은 동맥과 정맥을 통한 피의 흐름에 의해 생기는 열점을 인시하는 적외선 카메라를 사용한 열상 인식 방법과 3차원 측정기를 이용하여 눈, 코, 입, 콧구멍, 턱간의 각도와 거리, 뼈돌출 정도 등 얼굴의 특징점을 파악하는 방법으로 나뉨.

  -최근 얼굴 검출에는 유사하르 특징이 사용됨.

+ 2단계 : 누구의 얼굴인지 감별하거나 등록된 얼굴과 일치하는지 검사
  - 얼굴을 인식해 특정하기 위해서 얼굴의 주요 부분인 눈, 코,입의 거리와 모양을 이용해 판별함
  - 모자나 안경등의 장애물이 있어서 정확한 인식이 어려움으로 주성분 분석법이 많이 사용됨
  - 얼굴 이미지들에 대한 고유의 값을 구하여 고유성분 얼굴들을 추출하고, 이 값을 주어진 입력 얼굴 이미지의 값과 비교해 일치 여부를 판단함
  - 하지만 조명이나 표정변화에 따른 변화판단이 어려워 선형판별 분석이나 독립성분 분석 등 다양한 알고리즘 사용함

  - 플랫폼 관리형 인공지능 비서: 특정 플랫폼과 운영체제에 종속됨.
특정 플랫폼에서 생성되는 데이터와 실행되는 앱과 서비스를 모두 세밀하게 관리하고 조작할 수 있는 것이 장점이자 특징이나, 특정 플랫폼에 종속되어있기에 해당 플랫폼 외에서는 서비스를 이용하지 못한다는 단점이 있음 
ex) 애플 시리(iOS 및 맥OS 관리), MS 코타나(윈도우10 관리), 구글 나우(안드로이드 관리) 등
  - 오픈 API형 인공지능 비서:모든 플랫폼과 기기에서 이용할 수 있음
오픈 API를 이용해 스마트폰, 스마트 냉장고, 스마트티비, 컴퓨터 등 여러 플랫폼과 기기에 내장된 인공지능 서비스를 이용 가능한게 장점이자 특징으로 주로 사물 인터넷 기기를 제어하는데 활용됨. 특정 플랫폼과 기기의 데이터와 기능에 접근하기 위해 많은 노력이 필요하기 때문에 개발에 어려움이 많음. 또한 앱 형태로 서비스를 제공할 수도 있지만, 심도 있는 서비스를 제공하는데 어려움이 있음
ex) 아마존 알렉사, 구글 어시스턴트 등    

3. 사용자의 환경에 따른 종류   
  - 음성 대화형 사용자 환경 : 사용자와 인공지능 비서가 음성으로 정보를 주고 받음.
  - 예측형 사용자 환경 : 사용자의 데이터를 수집해 사용자의 명령을 미리 예측해서 비서 역할을 수행함.
  - 플랫폼은 음성대화형과 예측형 특징을 반반씩 갖추거나 음성 대화형 특징을 가짐.
  - 오픈API은 음성대화형을 갖추고 있는 경우가 많음.   

#####2. 원리 및 구조
인공지능 비서의 원리란 클라우드 데이터 센터 속에 인공신경망을 적용한 기본 인공지능 패키지(모델)를 설치한 후 감독학습과 비감독 학습이라는 두 가지 기계학습 기법을 통해 인공지능 비서의 처리 능력과 처리할 수 있는 분야를 확대시킨 것임. 또한 인공지능 비서는 자연어 처리 기술 즉, 사람의 말을 듣고 이를 이해할 수 있는 기능도 가지고 있음.인공지능 비서는 각 IT업체의 클라우드 데이터 속에 담겨 있어 사용자가 인터넷과 기기를 통해 접속해 인공지능 비서 서비스를 제공받기 때문에 인터넷 연결이 반드시 필요하며 주고 받는 데이터의 양도 많음. 사용자가 보유한 기기의 성능은 인공지능 비서의 성능과 관계가 없어 상대적으로 성능이 떨어지는 기기에서도 인공지능 비서 서비스를 쾌적하게 이용할 수 있음. 

#####3. 인공지능 스피커의 구성요소 
1. grammar - 사람 -> ai 스피커에게 하는 말.
  - grammar : 문법
사람이 인공지능 스피커에게 말하는 것으로 인공지능스피커 입장에서는 일종의 입력단위
즉, 인공지능 스피커가 사람의 목소리를 입력받는 요소를 의미함.
2. prompt - ai 스피커 -> 사람에게 하는 말.
  - prompt : (사람에게 어떤 결정을 내리도록어떤 일이 일어나도록) 하다[촉발하다]
prompt는 1. grammar 와 반대로 인공지능 스피커가 사람에게 말하는 것을 의미함
사람의 명령이나 질문한 것에 대한 대답, 또는 단순한 정보전달이 될 수 있음.
3. persona - 정체성
  - persona : (다른 사람들 눈에 비치는, 특히 그의 실제 성격과는 다른, 한 개인의) 모습인공지능 스피커가 갖는 성격/정체성을 의미함.
예를 들어 인공지능 비서의 고객층을 여자 비서를 원하는 사장님들이라고 가정시 persona를 20대 여성 비서 목소리, 말하는 내용 또한 사실 자료를 기반한 정확한 내용을 말하도록 설계해야함. 이처럼 persona에 따라 사용자와 관계성이 생김
4. prosody - 말투
prosody는 persona와 비슷한 개념으로, 예를 들면 목소리 높낮이, 음색 등이 포함될 수 있음. persona를 만들기 위한 개념과 같음.
5. interaction - 사람과 ai 스피커의 상호작용
사람과 ai 스피커의 상호작용을 말함.

#####4. 음성인식과정
알렉사를 사례로 음성 인식과정을 알아보면 다음과 같음
1. 목소리가 AI스피커의 마이크를 통해 입력되면 음성 신호를 음성 신호 벡터로 변환시킴.
2. 음성 신호 벡터는 단어 DB를 기반으로 분할
3. 분할된 음성 신호 벡터는 단어마다 고정된 길이의 하위 음성 신호 벡터인 언어 표현 정보로 추출  
4. 추출된 표현 정보는 입력된 명령에 알맞은 스킬을 결정
(스킬: 음악, 쇼핑등 해당 명령과 관련된 가장 상위 카테5. 고리일 수 있으며, 때로는 명령에 따라 복수의 스킬과 관련성을 가질 수 있음)
5. 언어 표현 정보를 통해 입력된 명령의 의도를 결정
6. 표현 정보에서 문장의 문법적 구조를 고려하여 의도/명령의 핵심 정보를 인지

## 5.학습목표

인공지능 기술에 대해 알아보면서 이미지 분석 방식에 대해 관심이 생겼다.
이미지 분석의 경우 일상 속에서 뿐만 아니라 헬스케어 분야에서도 활용도가 높다
실제로 구글에서는 구글 인공지능 헬스케어 팀이 만들어졌으며, 이 팀은 수백만개의 이미지를 라벨링한 후 이를 수백만번 반복적으로 학습시켜 암을 진단하는 알고리즘을 개발했다. 
이를 보고 이미지 분석과 헬스케어를 접목시켜 실생활에 유용한 알고리즘을 개발해보고 싶다. 이를 위해 인공지능 이미지 분석 API를 이용해 이미지 분석 알고리즘을 공부하며 어떻게 헬스케어와 접목시킬 수 있을지 연구해 볼 것이다.
